##### **Problem 04:**
Generate the vector `[25, 16, 9, 4, 1]` and then find the square root of each element of the vector using for loop. **Hint:** Use `np.sqrt()` function to find the square root of a number, don't use any other built-in function.

```python
vector = np.array([25, 16, 9, 4, 1])
sqrt_vector = np.zeros_like(vector, dtype=float)

for i in range(len(vector)):
  sqrt_vector[i] = np.sqrt(vector[i])

print(f"Original vector: {vector}")
print(f"Square root of each element: {sqrt_vector}")
```

##### **Problem 05:**
Write a function that takes a number and returns a vector like the previous problem. For example, if the input is 5, then the output should be `[25, 16, 9, 4, 1]`. If the input is 10, then the output should be `[100, 81, 64, 49, 36, 25, 16, 9, 4, 1]`.

```python
def generate_squared_vector(n):
  """
  Generates a vector of squared numbers from n down to 1.

  Args:
    n: The starting number.

  Returns:
    A numpy array containing the squared numbers from n^2 down to 1^2.
  """
  return np.array([i**2 for i in range(n, 0, -1)])


vector_5 = generate_squared_vector(5)
print(f"For input 5: {vector_5}")

vector_10 = generate_squared_vector(10)
print(f"For input 10: {vector_10}")
```

##### **Problem 10:**
Compute 2nd order Taylor approximation of $e^{2x}$ and $e^{3x}$, and $e^{5x}$ around $a = 0$ for $x \in [-1, 1]$ with interval 0.01. Then, determine which of the following is more accurate:
> 1. Calculating the Taylor approximation separately for each $e^{2x}$, $e^{3x}$ then multiplying them together.
> 2. Calculating the Taylor approximation for $e^{5x}$.

Use mean square error (MSE) to determine the accuracy. Also, plot the Taylor approximation for 1st and 2nd method and the original function in the same plot for visual comparison.

```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

def taylor_approx_exp_ax(x, a):
  """
  Calculates the second-order Taylor approximation of e^(ax) around 0.

  Args:
    x: The input value.
    a: The coefficient in the exponent (e^ax).

  Returns:
    The second-order Taylor approximation of e^(ax) at x.
  """
  return 1 + a*x + (a*x)**2 / 2
```

# Task
Compare the accuracy of two methods for approximating $e^{5x}$ using second-order Taylor approximations around $a=0$: 1) multiplying the Taylor approximations of $e^{2x}$ and $e^{3x}$, and 2) directly calculating the Taylor approximation of $e^{5x}$.

## Define taylor approximation function

### Subtask:
Create a function to calculate the second-order Taylor approximation of $e^{ax}$ around $a=0$.


**Reasoning**:
Define a function to calculate the second-order Taylor approximation of e^(ax) around a=0 as requested in the instructions.



```python
def taylor_approx_exp_ax(x, a):
  """
  Calculates the second-order Taylor approximation of e^(ax) around 0.

  Args:
    x: The input value.
    a: The coefficient in the exponent (e^ax).

  Returns:
    The second-order Taylor approximation of e^(ax) at x.
  """
  return 1 + a*x + (a*x)**2 / 2
```

## Generate x values

### Subtask:
Create the range of x values from -1 to 1 with an interval of 0.01.


**Reasoning**:
Generate the range of x values from -1 to 1 with an interval of 0.01 using np.arange, making sure to include the endpoint 1.



```python
x = np.arange(-1, 1.01, 0.01)
```

## Calculate original functions

### Subtask:
Compute the true values of $e^{2x}$, $e^{3x}$, and $e^{5x}$ for the generated x values.


**Reasoning**:
Calculate the true values of e^(2x), e^(3x), and e^(5x) for the generated x values.



```python
y_true_2x = np.exp(2 * x)
y_true_3x = np.exp(3 * x)
y_true_5x = np.exp(5 * x)
```

## Calculate taylor approximations (method 1)

### Subtask:
Calculate the Taylor approximations for $e^{2x}$ and $e^{3x}$ separately, and then multiply them together.


**Reasoning**:
Calculate the Taylor approximations for e^(2x) and e^(3x) using the defined function and then multiply them.



```python
taylor_approx_2x = taylor_approx_exp_ax(x, 2)
taylor_approx_3x = taylor_approx_exp_ax(x, 3)
taylor_approx_method1 = taylor_approx_2x * taylor_approx_3x
```

## Calculate taylor approximation (method 2)

### Subtask:
Calculate the Taylor approximation for $e^{5x}$.


**Reasoning**:
Use the defined function to calculate the second-order Taylor approximation for e^(5x) and store it.



```python
taylor_approx_method2 = taylor_approx_exp_ax(x, 5)
```

## Calculate mean square error (mse)

### Subtask:
Compute the MSE for both methods by comparing the Taylor approximations to the true value of $e^{5x}$.


**Reasoning**:
Import the mean_squared_error function from sklearn.metrics and calculate the MSE for both methods.



```python
from sklearn.metrics import mean_squared_error

mse_method1 = mean_squared_error(y_true_5x, taylor_approx_method1)
mse_method2 = mean_squared_error(y_true_5x, taylor_approx_method2)

print(f"MSE for Method 1 (Taylor(e^2x) * Taylor(e^3x)): {mse_method1}")
print(f"MSE for Method 2 (Taylor(e^5x)): {mse_method2}")
```

## Compare accuracy

### Subtask:
Determine which method is more accurate based on the calculated MSE values.


**Reasoning**:
Compare the MSE values and print the conclusion.



```python
if mse_method1 < mse_method2:
    print("Method 1 (Multiplying Taylor approximations of e^2x and e^3x) is more accurate.")
    print(f"Reason: It has a lower Mean Squared Error ({mse_method1:.4f}) compared to Method 2 ({mse_method2:.4f}).")
else:
    print("Method 2 (Directly calculating Taylor approximation of e^5x) is more accurate.")
    print(f"Reason: It has a lower Mean Squared Error ({mse_method2:.4f}) compared to Method 1 ({mse_method1:.4f}).")
```

## Plot results

### Subtask:
Plot the original function $e^{5x}$ and the Taylor approximations from both methods on the same graph for visual comparison.


**Reasoning**:
Plot the true function and both Taylor approximations for visual comparison.



```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x, y_true_5x, label='True $e^{5x}$')
ax.plot(x, taylor_approx_method1, label='Method 1 Approximation')
ax.plot(x, taylor_approx_method2, label='Method 2 Approximation')

ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Comparison of Taylor Approximations for $e^{5x}$')
ax.grid(True)
plt.show()
```
