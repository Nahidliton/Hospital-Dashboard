##### **Problem 04:**
Generate the vector `[25, 16, 9, 4, 1]` and then find the square root of each element of the vector using for loop. **Hint:** Use `np.sqrt()` function to find the square root of a number, don't use any other built-in function.

```python
vector = np.array([25, 16, 9, 4, 1])
sqrt_vector = np.zeros_like(vector, dtype=float)

for i in range(len(vector)):
  sqrt_vector[i] = np.sqrt(vector[i])

print(f"Original vector: {vector}")
print(f"Square root of each element: {sqrt_vector}")
```

##### **Problem 08:**
Plot $x + y + z = 1$ for $x, y \in [-3, 3]$ with interval 0.5. Use meshgrid to generate the grid. You can use any built-in function to plot the surface.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)

X, Y = np.meshgrid(x, y)
Z = 1 - X - Y

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Plot of $x + y + z = 1$')

plt.show()
```

##### **Problem 10:**
Compute 2nd order Taylor approximation of $e^{2x}$ and $e^{3x}$, and $e^{5x}$ around $a = 0$ for $x \in [-1, 1]$ with interval 0.01. Then, determine which of the following is more accurate:
> 1. Calculating the Taylor approximation separately for each $e^{2x}$, $e^{3x}$ then multiplying them together.
> 2. Calculating the Taylor approximation for $e^{5x}$.

Use mean square error (MSE) to determine the accuracy. Also, plot the Taylor approximation for 1st and 2nd method and the original function in the same plot for visual comparison.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Define the Taylor approximation function
def taylor_approx_exp(x, a, order):
    """Computes the Taylor approximation of e^u around 'a'."""
    u = x - a
    if order == 0:
        return np.exp(a)
    elif order == 1:
        return np.exp(a) * (1 + u)
    elif order == 2:
        return np.exp(a) * (1 + u + u**2 / 2)
    else:
        raise ValueError("Only orders 0, 1, and 2 are supported.")

# Define the range and interval for x
x = np.arange(-1, 1.01, 0.01)
a = 0 # Approximation point

# Original functions
y_e2x = np.exp(2 * x)
y_e3x = np.exp(3 * x)
y_e5x = np.exp(5 * x)

# Method 1: Taylor approximation of e^2x and e^3x separately, then multiply
taylor_e2x = taylor_approx_exp(2 * x, 2 * a, 2)
taylor_e3x = taylor_approx_exp(3 * x, 3 * a, 2)
method1_approx = taylor_e2x * taylor_e3x

# Method 2: Taylor approximation of e^5x
method2_approx = taylor_approx_exp(5 * x, 5 * a, 2)

# Calculate MSE for both methods
mse_method1 = mean_squared_error(y_e5x, method1_approx)
mse_method2 = mean_squared_error(y_e5x, method2_approx)

print(f"MSE for Method 1 (Taylor of e^2x * Taylor of e^3x): {mse_method1}")
print(f"MSE for Method 2 (Taylor of e^5x): {mse_method2}")

# Determine which method is more accurate
if mse_method1 < mse_method2:
    print("Method 1 is more accurate.")
elif mse_method2 < mse_method1:
    print("Method 2 is more accurate.")
else:
    print("Both methods have the same accuracy.")

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(x, y_e5x, label='$e^{5x}$ (Original)')
plt.plot(x, method1_approx, label='Method 1 (Taylor $e^{2x} \cdot$ Taylor $e^{3x}$)')
plt.plot(x, method2_approx, label='Method 2 (Taylor $e^{5x}$)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Taylor Approximation of $e^{5x}$')
plt.legend()
plt.grid(True)
plt.show()
```
